1. Explain the difference between Relative , Absolute and Mixed Cell Referencing.

ans - . Relative Cell Referencing

 A reference that changes automatically when a formula is copied to another cell.

Example: =A1+B1

  
   . Absolute Cell Referencing

Definition: A reference that remains fixed even when the formula is copied.

Symbol Used: $

Example:  =$A$1+$B$1

3. Mixed Cell Referencing

Definition: A reference where either the row or the column is fixed.

Examples:

$A1 → Column A is fixed, row changes

A$1 → Row 1 is fixed, column changes
